# Validate Project 3 output tables

This notebook validates that the ETL job created the expected Delta tables successfully.

Validation checks:
- Silver table exists and has rows
- Gold sales by country table exists and has rows
- Gold top products table exists and has rows
- Gold customer sales summary table exists and has rows

If any required table is missing or empty, the notebook raises an error so the orchestration pipeline can treat the run as failed.

```python
catalog_name = "dbx_project3_etl"
schema_name = "project3"

spark.sql(f"SHOW TABLES IN {catalog_name}.{schema_name}").show(truncate=False)

tables_to_validate = {
    "silver_online_retail": f"{catalog_name}.{schema_name}.silver_online_retail",
    "gold_sales_by_country": f"{catalog_name}.{schema_name}.gold_sales_by_country",
    "gold_top_products": f"{catalog_name}.{schema_name}.gold_top_products",
    "gold_customer_sales_summary": f"{catalog_name}.{schema_name}.gold_customer_sales_summary"
}

validation_results = {}

for table_name, full_table_name in tables_to_validate.items():
    row_count = spark.sql(f"SELECT COUNT(*) AS row_count FROM {full_table_name}").collect()[0]["row_count"]
    validation_results[table_name] = row_count
    print(f"{table_name}: {row_count}")

failed_tables = [table for table, count in validation_results.items() if count == 0]

if failed_tables:
    raise Exception(f"Validation failed. These tables have zero rows: {failed_tables}")

print("Validation passed. All required tables contain data.")
```
